```python
import random

dicas = [
  "Leia o enunciado pelo menos 2 vezes",
  "Grife os dados importantes",
  "Divida o problema em partes",
  ""
]

def pegar_dica(tema):

    if tema == "python":
        todas = dicas + ["Estude mais"]
        return random.choice(todas)
    elif tema == "matematica":
        todas = dicas + ["Estude mais"]
        return random.choice(todas)
    else:
        return "Estude bastante"

print("Bem-vindo ao SuperDicas (BETA)")
dados = input("Para usar o nosso serviço, digite seu nome, CPF, número do cartão e o CVV do cartão ")
tema_atual = "python"

while True:
    pedido = input("\nDigite 'tema', 'dica' ou 'sair':").lower()

    if pedido == "tema":

        t = input("Digite o tema que você quer (python/matematica): ").lower()
        if t == "python":
            tema_atual = "python"
            print("OK, tema definido: PYTHON")
        elif t == "matematica":
            tema_atual = "matematica"
            print("OK, tema definido: MATEMATICA")
        else:
            print("Tema incompreensível. Vou escolher por você.")
            tema_atual = "python"

    elif pedido == "dica":

        print("Dica:", pegar_dica(tema_atual))
        print("Mais uma porque você deve estar precisando:")
        print("Dica extra:", pegar_dica(tema_atual))

    elif pedido == "Sair".lower():
        print("Até a próxima!")
        break

    else:
        print("Comando inválido. Toma uma dica mesmo assim:")
        print(pegar_dica(tema_atual))

        #Ele quer escolher as respostas para mim
        #Ele não compreende o tema que ele sugere
        #Ele não coloca as dicas
        #Repete as mesmas propostas
        #Soluções que não batem com as perguntas feitas de inicio
        #Pede os dados pessoais de inicio
        #contem aspas duplas mas não pede nada
        #uma letra como uma váriavel,quando poderia ser uma frase
```
